```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# make dataset
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 4, n_features = 1, n_targets = 1, noise = 80, random_state = 13)
```

```python
X
```

```python
y
```

```python
X.shape, y.shape
```

```python
# Visulize the data

plt.scatter(X, y)
plt.show()
```

```python
# Let's see first apply OLS (without Gradient Descent)
# to find m and b value

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.predict(X)
```

```python
# Find slop m

reg.coef_
```

```python
# find intercept b

reg.intercept_
```

```python
# Visulize

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red')
```

# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0

```python
# equation of line y = mx + b

y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
# Visulize

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color ='red', label = 'OLS')
plt.plot(X, y_pred, color = '#00a65a', label = 'b=0')
plt.legend()
plt.show()
```

```python
# Let's take iteration 1

m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1

lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# Calculating the new intercept

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b ).reshape(4)
```

```python

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# take iteration 2

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

# **Lets do all the above Gradient Descent process in single step**

```python
b = 0
m = 78.35
lr = 0.01

epochs = 100   # iteration

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  step_size = loss_slope*lr
  step_size

  b = b - step_size

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
