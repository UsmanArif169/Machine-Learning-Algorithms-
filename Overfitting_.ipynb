# Overfitting

Overfitting happens when:

* Model learns training data too well
* But performs poorly on unseen data, usually test data

Training Error  → VERY LOW
Test Error      → HIGH

Compare performance on TRAIN vs TEST data -> If there is a big gap → overfitting.

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from slearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
data = load_diabetes()

X = data.data
y = data.target
```

```python
# I intentionally took 50% data for testing to make it overfit.

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42
)
```

```python
# We intentionally make the model too complex.
overfit_model = Pipeline([
    ("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2)),  # VERY HIGH DEGREE
    ("linear", LinearRegression())
])
```

```python
overfit_model.fit(X_train, y_train)
```

```python
y_train_pred = overfit_model.predict(X_train)
y_test_pred = overfit_model.predict(X_test)
```

```python
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print("Train R2:", train_r2)
print("Test R2 :", test_r2)
```

Method: Cross validation score


Overfitting Signal:

Very high train score and Lower CV score

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(overfit_model, X, y, cv=5, scoring="r2")

print("scores:", scores)

mean_score = scores.mean()
print("Train score    :", train_r2)
print("CV score (Mean):", mean_score)


if train_r2 - mean_score > 0.10:
    print("Overfitting"
```
